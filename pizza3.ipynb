```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('D:\\data.csv')
```

```python
df.head()
```

```python
catogorical_data =df.select_dtypes(object)
```

```python
numerical_data = df.drop(columns=['concept_type','city','region','target'])
```

```python
numerical_data
```

## Label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in catogorical_data:
    le = LabelEncoder()
    catogorical_data[col]=le.fit_transform(catogorical_data[col])
```

```python
catogorical_data
```

```python
X = np.concatenate((numerical_data,catogorical_data),axis=1)
```

```python
X.shape
```

```python
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.2 , random_state = 100)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print("LR Train MSE: ",lr_train_mse)
print("LR Train R2: ",lr_train_r2)
print("LR Test MSE: ",lr_test_mse)
print("LR Test R2: ",lr_test_r2)
```
